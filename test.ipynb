#### 1-3 2차 교호작용항까지 고려한 회귀분석 수행 및 변수 선택 과정 제시 

?교호작용이란
A의 효과가 B의 서로 다른 수준 B1과 B2에서 일관성 있게 나타나면 두 인자 A,B 간 교호작용 없다고함  
여기서 2차교호작용은 변수 3개까지 곱하는 조합 ? 1차교호작용 1x1 , 2차교호작용 1x1x1 
제곱은 곱하는게 아니라 다항의 조합  

```python
from itertools import permutations
comb = list(permutations(X_train.columns,3))
len(comb)
```

```python
comb
```

```python
variables= '+ '.join(list(X_train.columns)) +'+'+ '+'.join([':'.join(list(y)) for y in comb])
```

```python
variables
```

```python
from statsmodels.formula.api import ols

res = ols(f'Price~{variables}',data = pd.concat([X_train,y_train],axis=1)).fit()
res.summary()
```
